# Frequency Distribution


Frequency Distribution is used to find the frequency of each word occurring in a document.

```python
import nltk
```

We will be tokenizing our document into individual sentences and further break them down into words using `sent_tokenize` and `word_tokenize`.

```python
from nltk.tokenize import word_tokenize
```

The following sample text is extracted from the corpus of NLTK.

```python
my_file = open("2001-Bush.txt")
my_file_str = my_file.read()
```

## Word Tokenization

```python
words = word_tokenize(my_file_str)
print(words)
```

## Frequency Distribution

```python
fd = nltk.FreqDist(words)
fd.plot()
```

There are so many words in the plot that the plot is not easy to visualize. Therefore, `most_common(20)` helps us identify the 20 most common words in the document. The obvious thing to note is that stop words constitute the majority.

```python
freq = fd.most_common(20)
```
