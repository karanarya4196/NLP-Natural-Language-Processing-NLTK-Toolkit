# Bigrams, Trigrams and Ngrams

```python
import nltk
```

```python
my_file = open("2001-Bush.txt")
my_file_str = my_file.read()
```

## Word Tokenization

```python
words = nltk.word_tokenize(my_file_str)
```

## Bigrams

```python
output = list(nltk.bigrams(words))
print(output)
```

## Trigrams

```python
output = list(nltk.trigrams(words))
print(output)
```

## Ngrams

```python
from nltk import ngrams
output = list(ngrams(words, 6))
print(output)
```
